#### Database Engineering Script

```python
# dependencies

from sqlalchemy import Column, Integer, String, Float, Text, Date
from sqlalchemy.ext.declarative import declarative_base

Base = declarative_base()

from sqlalchemy import create_engine, func
from sqlalchemy.orm import Session

import pymysql
pymysql.install_as_MySQLdb()

import pandas as pd
```

```python
# read Stations CSV
stations_df = pd.read_csv("resources/original/hawaii_stations.csv")
stations_df
```

```python
# read first 20 from clean measurements CSV
measurements_df = pd.read_csv("clean_hawaii_measurements.csv")
measurements_df.head(20)
```

```python
# create Stations table
class Stations(Base):
    __tablename__ = "station_info"
    id = Column(Integer, primary_key = True)
    station = Column(String)
    name = Column(String)
    latitude = Column(Float)
    longitude = Column(Float)
    elevation = Column(Float)
```

```python
# create Measurements table
class Measurement(Base):
    __tablename__ = "climate_measurements"
    id = Column(Integer, primary_key = True)
    station = Column(String)
    date = Column(Integer)
    prcp = Column(Float)
    tobs = Column(Integer)
```

```python
# create SQL db and connect engine
engine = create_engine("sqlite:///hawaii.sqlite")
conn = engine.connect()

Base.metadata.create_all(conn)
```

```python
# send df data to SQL
station_data = stations_df.to_dict(orient = "records")
measurement_data = measurements_df.to_dict(orient = "records")
```

```python
conn.execute(Stations.__table__.delete())
conn.execute(Stations.__table__.insert(), station_data)
conn.execute(Measurement.__table__.delete())
conn.execute(Measurement.__table__.insert(), measurement_data)
```
